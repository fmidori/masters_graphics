```python
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline
```

```python
def stacked_bar(tsv,column):
    df = pd.read_csv(tsv, sep='\t')
    fig = plt.figure()
    sns.set_style('white')
    from matplotlib.colors import ListedColormap
    ax = df.set_index(column).T.plot(kind='bar', stacked=False,
              colormap=ListedColormap(sns.color_palette("Dark2",4)), #Dark2
              figsize=(20,9), #15,6.5
              fontsize=18,
              linewidth=2)
    plt.xlabel('Família', fontsize=18)
    plt.ylabel('Número', fontsize=18)
    #plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
    #plt.title("Seaborn Plot with Legend Outside")
    #plt.tight_layout()
    plt.setp(ax.get_legend().get_texts(), fontsize='20') # for legend text
    plt.setp(ax.get_legend().get_title(), fontsize='20') # for legend title
    #plt.legend(fontsize='x-large', title_fontsize='40')
    ax.set(yscale="log")
    #plt.savefig('toxins_to_avps.png', dpi=300)
    plt.show()
```

```python
stacked_bar('transcriptome_arach.tsv','Species')
```

```python
stacked_bar('proteome_arach.tsv','Species')
```

```python
stacked_bar('TicksPeptides.tsv','Legenda')
```

```python
stacked_bar('TickPeptidesClassTryp.tsv','Legenda')
```

```python
stacked_bar('TickPeptidesClassChemo.tsv','Legenda')
```
